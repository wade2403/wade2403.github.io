## Video 8 - Filtering in Pandas

```python
movies.head()
```

```python
booleans = []
for length in movies["duration"]:
    if length >= 200:
        booleans.append(True)
    else:
        booleans.append(False)
#This code helps us identify movies that are longer than 200 minutes
```

```python
booleans[0:5]
#The below shows that the third movie (GodFather pt 2) is longer than 200 minutes
```

```python
movies.shape
#Tells us number of rows and columns
```

```python
len(booleans)
#We want to see the same number of rows for this function
```

```python
is_long = pd.Series(booleans)
#This creates a new variable (is_long) as a pandas series (column)
```

```python
movies[is_long]
#This passes the is_long series into the movies dataframe
#THe below is now all the long movies
```

```python
#shorter way to filter and receive same results.  Inside the [] is the creation of the boolean variables by row
movies[movies.duration >= 200]
```

## Video 9 - Applying multiple filters

```python
#We want to look at long movies that are also of the genre drama
movies[(movies.duration >= 200) & (movies.genre == "Drama")]
```

```python
#The & requires both statements be true, we can replace this with an "or" statement
movies[(movies.duration >= 200) | (movies.genre == "Drama")]
```

```python
#We may also want to select multiple genre's using the isin function
movies[(movies.duration >= 200) & (movies.genre.isin(["Crime","Adventure"]))]
```

## Video 12 - Groupby function

```python
drinks.head()
```

```python
drinks.groupby("continent").beer_servings.mean()
#use group by when you have a "for each" statement
#Example: "for each" continent, what is the average beer servings?
```

```python
drinks.groupby("continent").beer_servings.agg(["min", "max", "mean", "count"])
#We can look at multiple descriptive stats using the .agg function
```

```python
drinks.groupby("continent").mean()
#Note: you dont have to specify a column when using groupby
```
